## Breast Cancer Analysis

#### Dataset Info:
Sample Data Set contains Breast cancer patient data. In breast cancer there are two types of tumor, Benign and Malignant. Data set is showing breast cancer patient has which type of tumor. 

```python
#import the required libraries
import numpy as np 
import pandas as pd
import sklearn.datasets
import seaborn as sns 
import matplotlib.pyplot as plt
```

#### Loading Data

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
data_frame.groupby('label').mean()
```

```python
data_frame['label'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("Count", labelpad=14)
plt.ylabel("Target Variable", labelpad=14)
plt.title("Count of TARGET Variable per category", y=1.02);
```

```python
100*data_frame['label'].value_counts()/len(data_frame['label'])
```

```python
# correlation
plt.figure(figsize=(12,12))
sns.heatmap(data_frame.corr(), cmap="Paired")
```
